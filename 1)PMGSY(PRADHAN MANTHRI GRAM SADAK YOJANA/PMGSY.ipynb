## Dataset Description

* The dataset is about the Pradhan Mantri Gram Sadak Yojana (PMGSY), a rural road development program in India. The PMGSY scheme was launched to provide all-weather road connectivity to unconnected villages and improve rural infrastructure. The dataset includes information on sanctioned and completed road and bridge projects under different phases of PMGSY, spanning various states and districts in India.This dataset can be used for infrastructure analysis, policy evaluation, and performance tracking of rural road projects in India


## Dataset Overview:

* The dataset is related to the Pradhan Mantri Gram Sadak Yojana (PMGSY), which focuses on rural road development in India. It contains 449 records and 14 columns that represent various metrics related to sanctioned and completed road and bridge works.

* Key Attributes and Their Descriptions

* STATE_NAME: The name of the state where the project is located.

* DISTRICT_NAME: The name of the district where the project is located.

* PMGSY_SCHEME: The phase of PMGSY (PMGSY-I, PMGSY-II, PMGSY-III).

* NO_OF_ROAD_WORK_SANCTIONED: The number of road projects sanctioned for construction.

* NO_OF_ROAD_WORKS_COMPLETED: The number of road projects successfully completed.

* NO_OF_ROAD_WORKS_BALANCE: The number of road projects that are still pending.

* NO_OF_BRIDGES_SANCTIONED: The number of bridges sanctioned under PMGSY.

* NO_OF_BRIDGES_COMPLETED: The number of bridges completed under PMGSY.

* NO_OF_BRIDGES_BALANCE: The number of bridges yet to be completed.

* LENGTH_OF_ROAD_WORK_SANCTIONED_KM: The total length of road work sanctioned (in kilometers).

* LENGTH_OF_ROAD_WORK_COMPLETED_KM: The total length of road work completed (in kilometers).

* LENGTH_OF_ROAD_WORK_BALANCE_KM: The remaining road length to be completed (in kilometers).

* COST_OF_WORKS_SANCTIONED_LAKHS: The total cost sanctioned for the projects (in lakh rupees).

* EXPENDITURE_OCCURED_LAKHS: The actual expenditure incurred for the projects (in lakh rupees).

   Initial Analysis of the Dataset  
* The dataset contains 449 records with 14 columns related to the PMGSY (Pradhan Mantri Gram Sadak Yojana) scheme. Key attributes include:
* State and District Names
* PMGSY Scheme Type (PMGSY-I, II, III, PM-JANMAN, RCPLWEA )
* Number of Road and Bridge Works (Sanctioned, Completed, Balance)
* Road Length and Cost (Sanctioned, Completed, Balance)
* Expenditure Incurred

```python
# Visualization: Road work completion rate per state
statewise_completion = df.groupby("STATE_NAME")[["NO_OF_ROAD_WORKS_COMPLETED", "NO_OF_ROAD_WORK_SANCTIONED"]].sum()
statewise_completion["COMPLETION_RATE"] = (statewise_completion["NO_OF_ROAD_WORKS_COMPLETED"] / statewise_completion["NO_OF_ROAD_WORK_SANCTIONED"]) * 100
```

```python
# Plot completion rates
statewise_completion["COMPLETION_RATE"].sort_values().plot(kind="bar",color = "orange")
plt.ylabel("Completion Rate (%)")
plt.title("State-wise Road Work Completion Rate")
plt.show()
```

```python
# Visualization: Expenditure vs Sanctioned Cost
sns.scatterplot(x=df["COST_OF_WORKS_SANCTIONED_LAKHS"], y=df["EXPENDITURE_OCCURED_LAKHS"], alpha=0.7,color = "green" )
plt.xlabel("Sanctioned Cost (Lakhs)")
plt.ylabel("Expenditure Incurred (Lakhs)")
plt.title("Expenditure vs. Sanctioned Cost")
plt.show()

```

```python
# Histogram showing the distribution of the number of sanctioned road works
sns.histplot(df["NO_OF_ROAD_WORK_SANCTIONED"], bins=30, kde=True,color = "purple")
plt.title("Distribution of Sanctioned Road Works")
plt.show()
```

```python
# Box plot to show the spread and outliers in sanctioned road length
sns.boxplot(x=df["LENGTH_OF_ROAD_WORK_SANCTIONED_KM"],color = "red")
plt.title("Box Plot of Sanctioned Road Length")
plt.show()
```

```python
# Scatter plot comparing sanctioned road length vs completed road length
sns.scatterplot(x=df["LENGTH_OF_ROAD_WORK_SANCTIONED_KM"], y=df["LENGTH_OF_ROAD_WORK_COMPLETED_KM"],color = "yellow")
plt.show()
```

```python
# Pie Chart: Breakdown of Projects (Sanctioned vs. Completed)
total_road_sanctioned = df["NO_OF_ROAD_WORK_SANCTIONED"].sum()
total_road_completed = df["NO_OF_ROAD_WORKS_COMPLETED"].sum()
total_road_balance = df["NO_OF_ROAD_WORKS_BALANCE"].sum()

labels = ["Completed", "Balance"]
sizes = [total_road_completed, total_road_balance]
colors = ["#4CAF50", "#FF9999"]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Pie Chart: Road Projects (Completed vs. Balance)")
plt.show()
```

```python
# Line plot to show the relationship between sanctioned cost and expenditure
sns.lineplot(x=df["COST_OF_WORKS_SANCTIONED_LAKHS"], y=df["EXPENDITURE_OCCURED_LAKHS"],color = "purple")
plt.title("Expenditure vs Sanctioned Cost")
plt.show()
```

```python
# Boxen plot to show distribution of expenditure by state
sns.lineplot(y=df["EXPENDITURE_OCCURED_LAKHS"], x=df["STATE_NAME"],color="red")
plt.title("Expenditure by State")
plt.show()
```

```python
# Count plot showing the frequency of different PMGSY schemes
sns.countplot(x=df["PMGSY_SCHEME"],color = "yellow")
plt.title("Count of PMGSY Schemes")
plt.show()
```

```python
# Pairplot for exploring relationships between numerical variables
sns.pairplot(df[["NO_OF_ROAD_WORK_SANCTIONED", "NO_OF_ROAD_WORKS_COMPLETED", "LENGTH_OF_ROAD_WORK_SANCTIONED_KM", "COST_OF_WORKS_SANCTIONED_LAKHS"]])
plt.show()
```

```python
# Selecting only numeric columns for the correlation heatmap
numeric_df = df.select_dtypes(include=["number"])

# Heatmap showing correlations between numerical features
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Strip plot showing the number of bridges completed by each state
sns.stripplot(x=df["STATE_NAME"], y=df["NO_OF_BRIDGES_COMPLETED"], jitter=True, size=3,color = "brown")
plt.title("Bridges Completed by State")
plt.show()
```

```python
# Strip plot comparing the number of sanctioned vs completed road works
plt.figure(figsize=(12, 6))
sns.stripplot(x=df["NO_OF_ROAD_WORK_SANCTIONED"], y=df["NO_OF_ROAD_WORKS_COMPLETED"], jitter=True, size=3,color = "purple")
plt.title("Strip Plot: Sanctioned vs Completed Road Works")
plt.show()
```

```python
# Pie chart showing distribution of sanctioned road works per PMGSY scheme
df.groupby("PMGSY_SCHEME")["NO_OF_ROAD_WORK_SANCTIONED"].sum().plot.pie(autopct="%1.1f%%")
plt.title("Sanctioned Road Works by Scheme")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
# Regression plot to analyze the relationship between cost and expenditure
sns.regplot(x=df["COST_OF_WORKS_SANCTIONED_LAKHS"], y=df["EXPENDITURE_OCCURED_LAKHS"],color = "green")
plt.title("Regression Plot: Cost vs Expenditure")
plt.show()
```

```python
# Line plot showing the number of road works completed in each district
df.groupby("DISTRICT_NAME")["NO_OF_ROAD_WORKS_COMPLETED"].sum().sort_values().plot(kind="line",color = "yellow")
plt.title("Road Works Completed by District")
plt.show()
```

```python
# Histogram showing the distribution of balance road length
sns.histplot(df["LENGTH_OF_ROAD_WORK_BALANCE_KM"], bins=30, kde=True,color="blue")
plt.title("Histogram of Balance Road Length")
plt.show()
```

```python
# Pie chart showing distribution of sanctioned bridges per PMGSY scheme
df.groupby("PMGSY_SCHEME")["NO_OF_BRIDGES_SANCTIONED"].sum().plot.pie(autopct="%1.1f%%")
plt.title("Bridges Sanctioned by Scheme")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
# KDE plot to visualize the distribution of remaining road works
sns.kdeplot(df["NO_OF_ROAD_WORKS_BALANCE"], fill=True,color = "orange")
plt.title("KDE Plot of Road Work Balance")
plt.show()
```

```python
# Kernel Density Estimation (KDE) plot for sanctioned costs
sns.kdeplot(df["COST_OF_WORKS_SANCTIONED_LAKHS"], fill=True,color = "purple")
plt.title("KDE Plot of Sanctioned Costs")
plt.show()
```

## Observations from Data Analysis

* Completion Rate Variability:
The completion rate of road works varies significantly across states and districts. Some states have nearly 100% completion, while others have a large backlog.

* Sanctioned vs. Completed Work:
The number of sanctioned road and bridge projects is generally higher than the completed projects, indicating delays or ongoing work. There are some regions where no road work has been completed despite approvals.

* Expenditure vs. Sanctioned Cost:
There is a positive correlation between sanctioned costs and actual expenditure, but in some cases, expenditures are lower than expected. This may indicate underutilization of funds or project delays.

* State-wise Differences:
Some states have a higher number of sanctioned projects but a lower completion rate, while other states efficiently complete most of the sanctioned projects.

* Project Balance (Pending Work):
Many districts have road and bridge projects that remain incomplete, leading to delays in improving rural connectivity. The balance length of roads yet to be completed varies from 0 km to nearly 500 km in some regions.

* Scheme-Wise Distribution:
PMGSY-I appears to have more sanctioned projects than PMGSY-II and PMGSY-III. The newer phases (PMGSY-II and III) may still be in the early execution phase.

* Potential Areas for Further Analysis

* State & District-wise Ranking: Identify which states and districts perform best and worst in terms of project completion.

* Funding Efficiency: Check if states that spend closer to the sanctioned budget complete more projects.

* Delays & Backlogs: Analyze why some states have a backlog and what factors contribute to it.
