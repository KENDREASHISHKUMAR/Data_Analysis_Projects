# Dataset Description
* The dataset falls within the Education and Academic Performance domain, specifically focusing on student assessment data. It is related to mid-term examination scores of students enrolled in an academic institution.
* The dataset contains 718 rows and 8 columns, but some columns have missing values. Below is a breakdown of each column:

* S.NO (Serial Number) - Numeric identifier for each record, but has 601 non-null values.
* SECTION - Represents the section name (e.g., "ALPHA"), with 691 non-null values.
* DV - Subject score (possibly Digital VLSI or another subject).
* M-II - Subject score (possibly Mathematics-II).
* PP - Subject score (possibly Probability and Probability).
* BEEE - Subject score (possibly Basic Electrical and Electronics Engineering).
* FL - Subject score (possibly Formal Languages or a similar subject).
* FIMS - Subject score (possibly Financial and Information Management Systems).

# Returns the number of students per each group

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
df['M2'] = pd.to_numeric(df['M2'], errors='coerce')
df['PP'] = pd.to_numeric(df['PP'], errors='coerce')
df['BEEE'] = pd.to_numeric(df['BEEE'], errors='coerce')
df['FL'] = pd.to_numeric(df['FL'], errors='coerce')
df['FIMS'] = pd.to_numeric(df['FIMS'], errors='coerce')
df['TOTAL'] = df['DV'] + df['M2'] + df['PP'] + df['BEEE'] + df['FL'] + df['FIMS']
df
```

```python
df["GRADE"]=((df["TOTAL"]/120)*10).round()
df
```

# Here we create a new coloum GRADE which shows the GPA of each student respectively 

```python
def assign_grade(PERCENTAGE):
    if PERCENTAGE >= 90:
        return 'A'
    elif PERCENTAGE >= 80:
        return 'B'
    elif PERCENTAGE >= 70:
        return 'C'
    elif PERCENTAGE >= 60:
        return 'D'
    else:
        return 'F'

df['GRADE'] = df['PERCENTAGE'].apply(assign_grade)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 2
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)
calculated_chi_square = 3.747

x = np.linspace(0, 10, 1000)
y = stats.chi2.pdf(x, df)

plt.plot(x, y, label='Chi-Square Distribution (df=2)')
plt.fill_between(x, y, where=(x > critical_value), color='blue', alpha=0.5, label='Critical Region')
plt.axvline(calculated_chi_square, color='red', linestyle='dashed', label='Calculated Chi-Square')
plt.axvline(critical_value, color='purple', linestyle='dashed', label='Critical Value')
plt.title('Chi-Square Distribution and Critical Region')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density Function')
plt.legend()
plt.show()
```

## Observations from Data Analysis

* **Column Renaming:** "M-II" was renamed to "M2" for simplification.
* **New Columns Created:**
  - "Total": Sum of marks across subjects.
  - "Percentage": Derived from total marks to indicate student performance.
* **Data Type Conversions:**
  - All values converted to integers for effective analysis.
* **Grading System:**
  - Introduced "GRADE" column, later upgraded to letter grades (A, B, C, D, F).

* **General Performance Trend:**
  - Most students scored **80-100%**, indicating high performance.
  - A few students had **significantly lower scores**, suggesting struggles in certain subjects.

* **Subject-Wise Performance:**
  - **DV :** Most students scored **17-20**.
  - **M2 :** Performance was **polarized**, with scores either **very high (17.5-20) or very low (0-2.5)**.
  - **BEEE:** Majority scored **17.5-20**.
  - **PP :** Most students performed well, with **18-20 marks**.
  - **FL :** Highest **mean marks** among all subjects.
  - **FIMS :** Performance ranged **10-17.5 marks**.

* **Pie Chart Analysis:**
  - **FL had the highest (19.3%)** of total marks.
  - **M2 had the lowest (12.4%)**, indicating weak performance overall.
  - **Max Marks Distribution:** Equal contribution of **16.7%** across subjects, except for M2.

* **Correlation Analysis:**
  - **FL showed a strong positive correlation** with Total Marks.
  - **M2 had the weakest correlation**, making it an outlier in performance trends.

* **High Scorers (110-120 Total Marks):**
  - **DV:** Majority scored **18-20**.
  - **M2:** Most scored **17-20**, with **11 students scoring 20**.
  - **PP:** Many scored **19-20**, with the lowest being **11**.
  - **BEEE:** Maximum **20 by 11 students**, minimum **15 by one student**.
  - **FL:** Most students scored **18-20**, reinforcing strong correlation with Total Marks.

* **Low Scorers (0-60 Total Marks):**
  - **DV:** Majority scored between **5 and 15**.
  - **FL & PP:** Students with low total marks generally performed poorly in these subjects as well.

* **Final Observations & Recommendations:**
  - **FL (Formal Languages) is a strong subject**, with consistently high performance.
  - **M2 (Mathematics-II) remains a challenge**, with many students either excelling or failing.
  - **Backlogs Analysis:** Further investigation needed into subjects with high failure rates.

* **Performance Groups:**
  - **Top Performers:** Consistently score **18+** in all subjects.
  - **Struggling Students:** Score low in **M2, DV, and PP**.
